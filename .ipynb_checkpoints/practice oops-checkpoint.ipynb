```python
# Create base class Shape with methods to calculate area and perimeter and derive classes circle and square.



import math

class Shape:
    
    def calculate_area(self):
        pass

   
    def calculate_perimeter(self):
        pass
        


class Circle(Shape):
    
    def __init__(self, radius):
        self.radius = radius
  
    def calculate_area(self):
        return math.pi * self.radius**2
       

    def calculate_perimeter(self):
        return 2 * math.pi * self.radius
        

class Square(Shape):

     def __init__(self, side):
         self.side = side

     def calculate_area(self):
         return self.side ** 2
       

     def calculate_perimeter(self):
        return 4 * self.side
       

r = 7
circle = Circle(r)
circle_area = circle.calculate_area()
circle_perimeter = circle.calculate_perimeter()

print("Radius of the circle:", r)
print("Circle Area:", circle_area)
print("Circle Perimeter:", circle_perimeter)
print("\n---------------------------------------------------------------------\n")

s = 7
square = Square(s)
square_area = square.calculate_area()
square_perimeter = square.calculate_perimeter()

print("Side of the square:", s)
print("Square Area:", square_area)
print("Square Perimeter:", square_perimeter)



```

```python
# Implement a hierarchy to represent different types of employees(Manager,Engineer) with their attributes:

class Employee:
    def __init__(self,id,name,gender,city,salary):
        self.id = id
        self.name = name
        self.gender = gender
        self.city = city
        self.salary = salary

class Manager(Employee):
    def __init__(self,id,name,gender,city,salary,role):
        super().__init__(id,name,gender,city,salary)
        self.role = role

    def man_details(self):
        print(f'Id is : {self.id}')
        print(f'Name is : {self.name}')
        print(f'Gender is : {self.gender}')
        print(f'City is : {self.city}')
        print(f'Salary is : {self.salary}')
        print(f'Role is : {self.role}')

class Engineer(Employee):
    def __init__(self,id,name,gender,city,salary,en_role):
        super().__init__(id,name,gender,city,salary)
        self.en_role = en_role

    def details(self):
        print(f'Id is : {self.id}')
        print(f'Name is : {self.name}')
        print(f'Gender is : {self.gender}')
        print(f'City is : {self.city}')
        print(f'Salary is : {self.salary}')
        print(f'Role is : {self.en_role}')

manager = Manager(101,"Rohit","Male","Mumbai",1000000,"Management")
manager.man_details()

print("\n------------------------------------------------------------------------\n")

engineer = Engineer(102,"Mohit","Male","Chennai",2637354,"Engineering role")
engineer.details()
    
```

```python
# Write a Python program that uses inheritance to represent a hierarchy of shapes(Triangle, Rectangle etc)

class Shape:
    
    def calculate_area(self):
        pass

   
class Triangle(Shape):

    def __init__(self,height,base):
        self.height = height
        self.base = base

    def calculate_area(self):
        area = 0.5 * self.height * self.base
        print("Area of triangle is : ",area)

class Rectangle(Shape):
    
    def __init__(self,length,width):
        self.length = length
        self.width = width
         
    def calculate_area(self):
        area = self.length * self.width
        print("Area of Rectangle is : ",area)

h = int(input("Enter height/length :"))
b = int(input("Enter base/width :"))
print("\n---------------------------------------------\n")

tri = Triangle(h,b)
tri.calculate_area()
print("\n---------------------------------------------\n")

rect = Rectangle(h,b)
rect.calculate_area()
print("\n---------------------------------------------\n")
```

```python
# Create a class hierarchy to represent different types of animals(Bird,Fish) with their own attributes and methods.

class Animals:

    def features(self):
        pass

class Bird(Animals):

    def __init__(self,name,color,scientific_name,kingdom):
        self.name = name
        self.color = color
        self.scientific_name = scientific_name
        self.kingdom = kingdom

    def features(self):
        print(f'Name of bird is : {self.name}')
        print(f'Color of bird is : {self.color}')
        print(f'Scientific of bird is : {self.scientific_name}')
        print(f'Kingdom of bird is : {self.kingdom}')


class Fish(Animals):

    def __init__(self,name,color,scientific_name,class1,kingdom):
        self.name = name
        self.color = color
        self.scientific_name = scientific_name
        self.class1 = class1
        self.kingdom = kingdom

    def features(self):
        print(f'Name of fish is : {self.name}')
        print(f'Color of fish is : {self.color}')
        print(f'Scientific of fish is : {self.scientific_name}')
        print(f'CLass name of fish is : {self.class1}')
        print(f'Kingdom of bird is : {self.kingdom}')

bird = Bird("Parrot","Parrot Green","Pscittaciformes","Animalia")
bird.features()
print("\n---------------------------------------------\n")

fish = Fish("Golden Fish","Golden","Carassius auratus","Pisces","Animalia")
fish.features()
```
