```python
class Calculator:
    def __init__(self):
        self.numbers = []

    def accept(self, *args):
        self.numbers.extend(args)

    def addition(self):
        x = sum(self.numbers)
        print("-----------------------------------------")
        print("Addition of numbers is : ", x)

    def subtraction(self):
        diff = self.numbers[0]    
        for each in range(1, len(self.numbers)):     
            diff = diff - self.numbers[each]
        print("-----------------------------------------")
        print("Subtraction of numbers is : ", diff)

    def multiplication(self):
        mult = self.numbers[0]
        for each in range(1, len(self.numbers)):
            mult = mult * self.numbers[each]
        print("-----------------------------------------")
        print("Multiplication of numbers is : ", mult)

    def division(self):
        div = self.numbers[0]
        for each in range(1, len(self.numbers)):
            if self.numbers[each] != 0:
                div = div / self.numbers[each]
            else:
                print("Error: Division by zero is not allowed.")
                return
        print("-----------------------------------------")
        print("Division of numbers is : ", div)

    def modulus(self):
        mod = self.numbers[0]
        for each in range(1, len(self.numbers)):
            mod = mod % self.numbers[each]
        print("-----------------------------------------")
        print("Modulus of numbers is : ", mod)

    def exponent(self):
        exp = self.numbers[0]
        for each in range(1, len(self.numbers)):
            exp = exp ** self.numbers[each]
        print("-----------------------------------------")
        print("Exponent of numbers is : ", exp)

    def remainder(self):
        rem = self.numbers[0]
        for each in range(1, len(self.numbers)):
            rem = rem // self.numbers[each]
        print("-----------------------------------------")
        print("Remainder of numbers is : ", rem)

    def details(self):
        print("""
        \n-------------------------------------------
        \nWhat do you want to perform:
        1.Addition
        2.Subtraction
        3.Multiplication
        4.Division
        5.Modulus
        6.Exponent
        7.Remainder
        8.Exit
        \n-------------------------------------------
        """)
        while True:
            try:
                print("1.Addition/2.Subtraction/3.Multiplication/4.Division/5.Modulus/6.Exponent/7.Remainder/")
                print("\n")
                option = int(input('Choose any option from above list: '))
                print("\n")
            except ValueError:
                print('Error: Enter a correct option.')
                continue
            else:
                if option == 8:
                    print("          ----------Thanks for using my Calculator-------------")
                    break
                while True:
                    user_input = input("Enter a number (Type 'ok' to stop and perform calculation): ")
                    if user_input.lower() == 'ok':
                        break
                    try:
                        num = float(user_input)
                        self.accept(num)
                    except ValueError:
                        print("Invalid input. Please enter a valid number or 'ok' to perform calculation.")
                if option == 1:
                    print("-----------------------------------------\n")
                    print("ADDITION :-")
                    self.addition()
                    print("-----------------------------------------\n")
                elif option == 2:
                    print("-----------------------------------------\n")
                    print("SUBTRACTION :-")
                    self.subtraction()
                    print("-----------------------------------------\n")
                elif option == 3:
                    print("-----------------------------------------\n")
                    print("MULTIPLICATION :-")
                    self.multiplication()
                    print("-----------------------------------------\n")
                elif option == 4:
                    print("-----------------------------------------\n")
                    print("DIVISION :-")
                    self.division()
                    print("-----------------------------------------\n")
                elif option == 5:
                    print("-----------------------------------------\n")
                    print("MODULUS :-")
                    self.modulus()
                    print("-----------------------------------------\n")
                elif option == 6:
                    print("-----------------------------------------\n")
                    print("EXPONENT :-")
                    self.exponent()
                    print("-----------------------------------------\n")
                elif option == 7:
                    print("-----------------------------------------\n")
                    print("REMAINDER :-")
                    self.remainder()
                    print("-----------------------------------------\n")
                else:
                    print('Error: Enter a correct option.')
                self.numbers = []

try:
    Calculator().details()
except:
    print('There might be some issue visit later')
```
