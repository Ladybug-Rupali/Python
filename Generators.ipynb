* __Generator are the functions that returns transversal object and a simple way of creating iterators.__
* __transvers the entire items at once.__
* __Can also be the expression having syntax similar to list comprehension.__
* __There is lot of complexity in creating iterators in pyton, it is required to implement iter() and next() method to kepp the track of internal state.__
* __Creating iterator is lengthy process therefore  generator makes it simple..__
* __If there is no value in iterations it raises stopIteration error.__
* __Its simple to create generator just def same as function and 'yeild' instead of return is used.__
* __Yeild is resposible to cintro; flow of generator function.__
* 

```python
def function():
    for i in range(10):
        if i % 2 == 0:
            yield i

num = []
for i in function():
    num.append(i)

print(num)
            
```

```python
def message():
    m_one = "Hello Rupali"
    yield m_one

    m_two = "Congratulations Riya"
    yield m_two

    m_three = "Python language"
    yield m_three

    m_four = "Fourth Varible"
    yield m_four

try:
    x = message()
    print(next(x))
    print(next(x))
    print(next(x))
    print(next(x))
    print(next(x))

except StopIteration:
    print('''
    data ended here!''')


```

```python
def multi_table(n):
    for i in range(1,11):
        yield i * n
        i += 1

emp_list = []
for i in multi_table(20):
    emp_list.append(i)
print(emp_list)

```
