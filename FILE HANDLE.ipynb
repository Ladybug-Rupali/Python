```python
import glob

class FileHandler:
    def __init__(self):
        self.drive = "C:\\Users\\prasa\\Downloads\\"
        self.file_name = ""
        self.file_found = False
        self.all_files = []

    def take_input(self):
        x = input("Enter the file name: ")
        self.file_name = self.drive + x
        try:
            with open(self.file_name, "r") as file:
                self.file_found = True
                print("File found!")
        except FileNotFoundError:
            self.file_found = False
            print("File not found!")

    def write_file(self):
        if self.file_found:
            content = input("Enter the content to write to the file: ")
            with open(self.file_name, "w") as file:
                file.write(content)
            print("File written successfully!")
        else:
            print("File not found!")

    def open_file(self):
        if self.file_found:
            with open(self.file_name, "r") as file:
                content = file.read()
                print("File content:")
                print(content)
        else:
            print("File not found!")

    def search_file_again(self):
        self.take_input()

    def show_all_files(self):
        self.all_files = glob.glob(self.drive + "*")
        print("All files in the directory:")
        for file in self.all_files:
            print(file.split("\\")[-1])

    def read_file(self):
        if self.file_found:
            with open(self.file_name, "r") as file:
                content = file.read()
                print("File content:")
                print(content)
        else:
            print("File not found!")

    def exit(self):
        print("Exiting the program. Goodbye!")

def main():
    file_handler = FileHandler()
    while True:
        print("\nOptions:")
        print("1. Take input from user as filename")
        print("2. Write to file")
        print("3. Open your file")
        print("4. Search file again")
        print("5. Show all files")
        print("6. Read file")
        print("7. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            file_handler.take_input()
        elif choice == "2":
            file_handler.write_file()
        elif choice == "3":
            file_handler.open_file()
        elif choice == "4":
            file_handler.search_file_again()
        elif choice == "5":
            file_handler.show_all_files()
        elif choice == "6":
            file_handler.read_file()
        elif choice == "7":
            file_handler.exit()
            break
        else:
            print("Invalid choice. Please try again!")

if __name__ == "__main__":
    main()
```

```python
class FileOpenCode:
    def __init__(self):
        self.drive = "C:\\Users\\prasa\\Downloads\\"
        self.file_name = ""

    def create_new_file(self):
        x = input("Enter the file name: ")
        self.file_name = self.drive + x + '.txt'
        n = int(input("Enter number of lines : "))

        with open(self.file_name, "w") as file:
             for i in range(1,n+1):
                 content = input(f"Input your line no {i} : ")
                 file.write(content)
                 file.write('\n')
             print("File created successfully!")

    def open_file(self):
        x = input("Enter the file name: ")
        self.file_name = self.drive + x + '.txt'
        try:
            with open(self.file_name, "r") as file:
                content = file.read()
                print("File content:")
                print(content)
        except FileNotFoundError:
            print("File not found!")

    def overwrite_file(self):
        x = input("Enter the file name: ")
        self.file_name = self.drive + x + '.txt'
        try:
            with open(self.file_name, "r") as file:
                print("File found!")
                choice = input("Do you want to overwrite it? (yes/no): ")
                if choice.lower() == "yes":
                    n = int(input("Enter number of lines : "))

                    with open(self.file_name, "w") as file:
                         for i in range(1,n+1):
                             content = input(f"Input your line no {i} : ")
                             file.write(content)
                             file.write('\n')
                         print("File overwritten successfully!")
                else:
                    print("File not overwritten.")
        except FileNotFoundError:
            print("File not found. Cannot overwrite a file that does not exist.")

    def search_file_again(self):
        x = input("Enter the file name: ")
        self.file_name = self.drive + x + '.txt'
        try:
            with open(self.file_name, "r") as file:
                print("File found!")
        except FileNotFoundError:
            print("File not found!")

    def exit(self):
        print("--------------------------------------Bye----------------------------------------")

    def details(self):
        print("\nOptions:")
        print("1. Create new file")
        print("2. Open file")
        print("3. Overwrite file")
        print("4. Search file again")
        print("5. Exit")
        
        while True:
            print("\n--------------------------------------------------\n")
            choice = input("Enter your choice: ")
            if choice == "1":
                self.create_new_file()
            elif choice == "2":
                self.open_file()
            elif choice == "3":
                self.overwrite_file()
            elif choice == "4":
                self.search_file_again()
            elif choice == "5":
                self.exit()
                break
            else:
                print("Invalid choice. Please try again!")

FileOpenCode().details()
```
